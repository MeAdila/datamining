```python
# kode data gender
data_cherbourg = "Male"
data_southampton = "Female"

# nilai binary
nilai_satu = 1
nilai_nol = 0

def ubah_biner(gender):
    return nilai_satu if gender == data_southampton else nilai_nol
```

```python
# CONSTAN VARIABLE
DECREMENT_BY_ONE = 1
INCREMENT_BY_ONE = 1

CONTINGENCY_TABLE_VALUE = {
    "q" : (1,1),
    "r" : (1,0),
    "s" : (0,1),
    "t" : (0,0),
}
```

```python
def get_series(al, idx, series):
    return al.loc[(idx), series]
```

```python
def get_dissimilarity_dataset(al, series_index = [], series = []):
    first_series = get_series(al, series_index[0], series)
    second_series = get_series(al, series_index[1], series)
    dataset = pd.concat([first_series,second_series],axis=1)
    return dataset.T
```

```python
def count_contingency_value(al, start_index = 0, last_index = 1):

    CONTINGENCY_VALUE = {
        "q" : 0,
        "r" : 0,
        "s" : 0,
        "t" : 0,
    }

    column_range = al.shape[1]

    for column in range(column_range):
        for value in CONTINGENCY_TABLE_VALUE:
            item = list((tuple(al.loc[(start_index):(last_index), al.columns[column]]) == CONTINGENCY_TABLE_VALUE[value], value))
            if item[0] == True:
                if item[1] == "q":
                    CONTINGENCY_VALUE["q"] += 1
                if item[1] == "r":
                    CONTINGENCY_VALUE["r"] += 1
                if item[1] == "s":
                    CONTINGENCY_VALUE["s"] += 1
                if item[1] == "t":
                    CONTINGENCY_VALUE["t"] += 1

    return CONTINGENCY_VALUE
```

```python
d_e_1_2 = count_contingency_value(dc_1_2, 1, 2)
```

```python
def measure_dissimilarity_binary_value_assymetric_distance(contingency_value):

    return (contingency_value["r"] + contingency_value["s"]) / (contingency_value["q"] + contingency_value["r"] + contingency_value["s"])
```

```python
d_1_2 = measure_dissimilarity_binary_value_assymetric_distance(d_e_1_2)
d_1_3 = measure_dissimilarity_binary_value_assymetric_distance(d_e_1_2)
d_1_4 = measure_dissimilarity_binary_value_assymetric_distance(d_e_1_2)
```
