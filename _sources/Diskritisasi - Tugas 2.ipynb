##**Metode**

  *   Diskritisasi dengan Equal Width dan Equal Frequency 
      *   Equal width : dikritisasi yang membagi data numerik menjadi beberapa kelompok dengan lebar kelompok yang kurang lebih sama besar. 
      *   Equal frequency : dikritisasi yang membagi data numerik menjadi beberapa kelompok dengan jumlah anggota per kelompok yang kurang lebih sama besar.
  *   Diskritisasi dengan basis Entropy

       Entropy digunakan untuk menentukan yang manakah node yang akan
menjadi pemecah data latih berikutnya

        
*  Rumus Entropy

$$
\operatorname{Entropy}\left(D_1\right)=-\sum_{i=1}^m P i \log _2 P i
$$

*  Rumus Gain

$$
\operatorname{Gain}\left(E_{\text {new }}\right)=E_{\text {initial }}-E_{\text {new }}
$$


*  Rumus Info
$$
\operatorname{Info}_A(D)=\frac{\left|D_1\right|}{|D|} \operatorname{Entropy}\left(D_1\right)+\frac{\left|D_2\right|}{|D|} \operatorname{Entropy}\left(D_2\right)
$$

###**Basis Entropy**

Menambahkan sempel data yang akan di hitung

```python
sample=al[['sepal_length']]
sample.describe()
```

Membuat kategori secara random untuk data sempel

```python
np.random.seed(0)
sample["kategori"] = np.where(np.random.choice(2,sample.shape[0]) < 1,"A","B")
sample
```

### Membuat fungsi GetOverollCategory (keseluruhan) yang digunakan untuk menghitung data keseluruhan yang nantinya digunakan untuk menghitung entropy

```python
def GetOverollCategory(col):
  group = sample.loc[:, :].groupby("kategori").count()
  a = group.loc['A', col]
  b = group.loc["B", col]
  return (a,b,a+b)
```

### Memasukan fungsi splitter yang digunakan untuk membuat split antara value yang telah ditentukan lalu mengembalikan data yang telah dipisahkan

```python
def splitter(value, col):
  #get data less and greater from value
  less = sample[sample[col] <= value]
  greater = sample[sample[col] > value]

  #calculate into category for each data
  less_group = less.loc[:, :].groupby("kategori").count()
  greater_group = greater.loc[:, :].groupby("kategori").count()

  #get value based on category
  less_category_A = less_group.loc['A',col]
  less_category_B = less_group.loc["B",col]
  greater_category_A = greater_group.loc['A',col]
  greater_category_B = greater_group.loc["B",col]

  return (
      [less_category_A, less_category_B, less_category_A + less_category_B],
      [greater_category_A, greater_category_B, greater_category_A + greater_category_B]
  )
```

### Membuat fungsi entropy untuk melakukan perhitungan entropy (menggunakan rumus entropy)

```python
def entropy (d):
  r1 = (d[0] / d[2]) * np.log2(d[0] / d[2])
  r2 = (d[1] / d[2]) * np.log2(d[1] / d[2])
  return np.sum([r1, r2]) * -1
```

### Membuat fungsi info

```python
def info(d):
  r1 = (d[0][2] / sample.shape[0]) * entropy(d[0])
  r2 = (d[1][2] / sample.shape[0]) * entropy(d[1])
  return r1 + r2
```

### Membuat fungsi gain

```python
def gain(Einitial, Enew):
  return Einitial - Enew
```

### Membuat Dinitial


```python
get_overoll = GetOverollCategory("sepal_length")
entropy_data = entropy(get_overoll)
print(get_overoll)
print(entropy_data)
```

### Melakukan beberapa split

```python
split1 = splitter(5.5, "sepal_length")
info_split1 = info(split1)
gain(entropy_data, info_split1)
```

```python
split2 = splitter(6.5, "sepal_length")
info_split2 = info(split2)
gain(entropy_data, info_split2)
```
